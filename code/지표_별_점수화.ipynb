### 상권변화 지표

- 2022년 4분기 데이터 사용
- LL(다이나믹):1, HL(상권축소):2, LH(상권확장):3 , HH(정체):4 으로 레이블 인코딩

```python
# 상권변화지표 (운영영업기간, 폐업영업기간)
commerce_2022 = pd.read_csv('/content/drive/MyDrive/광진구팟/Data/상권분석/2022상권변화지표.csv')
```

```python
commerce_2022
```

```python
commerce_2022['상권_변화_지표'] = commerce_2022['상권_변화_지표'].map({'LL':1, 'HL':2, 'LH':3, 'HH':4})
commerce_2022['행정동명'] = commerce_2022['행정동명'].map(lambda x: x.replace("제",""))
```

### 전체 매출

```python
sales_광진 = pd.read_csv('/content/drive/MyDrive/광진구팟/Data/상권분석/sales_광진.csv')
sales_광진
```

```python
sales_광진.columns
```

```python
# 상권별 매출액 합산 (2022년)
매출액 = sales_광진.groupby('상권_코드_명')[['상권_코드_명','분기당_매출_금액']].sum().rename(columns = {'분기당_매출_금액':'전체 매출'}).reset_index()
매출액
```

### 업종의 다양성

```python
업종_다양성 = sales_광진[['상권_코드_명','서비스_업종_코드_명']].drop_duplicates().groupby('상권_코드_명')['서비스_업종_코드_명'].count().reset_index().rename(columns = {'서비스_업종_코드_명':'업종_종류수'})
업종_다양성
```

### 지하철

```python
subway =  pd.read_csv('/content/drive/MyDrive/광진구팟/Data/접근성/지하철 final.csv')
subway.drop(['가까운_지하철역_명'], axis=1, inplace = True)
subway.rename(columns={'거리':'가까운 지하철역까지 거리', '650m_반경내_역':'650m_반경내_역수'}, inplace = True)
```

### 지표 별 점수화

```python
commerce_2022
```

```python
score = commerce_2022
score = pd.merge(score, rent[['행정동명','전체']], on = '행정동명')
```

```python
score.rename(columns = {'전체':'임대료_불안정성'}, inplace = True)
score['임대료_불안정성'] = 100/score['임대료_불안정성'] # 불안정성이 클수록 안좋으므로 역수를 취함.
```

```python
# 매출 데이터가 없는 자양종합시장(기능상실) 상권 제거
score.drop([20], inplace = True)
```

```python
score = pd.merge(score, 매출액, on = '상권_코드_명')
```

```python
score = pd.merge(score, 업종_다양성, on = '상권_코드_명')
```

```python
score = pd.merge(score, walk, on = '상권_코드_명')
```

```python
score = pd.merge(score, bus, on = '상권_코드_명')
```

```python
score.rename(columns = {'거리':'가까운 버스정류장까지 거리'}, inplace = True)
```

```python
score['가까운 버스정류장까지 거리'] = 100/score['가까운 버스정류장까지 거리'] # 버스정류장까지의 거리가 클수록 안좋으므로 역수를 취함.
```

```python
score = pd.merge(score, subway, on = '상권_코드_명')
```

```python
score['가까운 지하철역까지 거리'] = 100/score['가까운 지하철역까지 거리'] # 버스정류장까지의 거리가 클수록 안좋으므로 역수를 취함.
```
